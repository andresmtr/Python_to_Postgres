#### Create id to tables

```python
def create_id(columns):

    global df_0
    df_0 = pd.DataFrame(df[columns])
    df_0 = df_0.groupby([columns], as_index=False ,dropna=False).sum()
    df_0['id_'+columns] = df_0.index+1
```

```python

def merge_id(df_0, columns):
    
    global df
    df = df.merge(df_0, on=[columns], how='outer', suffixes=('', '_y'))
    df.drop(df.filter(regex='_y$').columns, axis=1, inplace=True)
    #return df
```

```python
create_id(columns='name_franchise')
merge_id(df_0, columns='name_franchise')

create_id(columns='name_game')
merge_id(df_0, columns='name_game')

create_id(columns='name_gender')
merge_id(df_0, columns='name_gender')

create_id(columns='name_platform')
merge_id(df_0, columns='name_platform')

create_id(columns='name_shop')
merge_id(df_0, columns='name_shop')

create_id(columns='name_type')
merge_id(df_0, columns='name_type')

create_id(columns='name_launcher')
merge_id(df_0, columns='name_launcher')

```

```python
## Verify the information to add

def single_insert(conn, insert_req):
    """ Execute a single INSERT request """
    cursor = conn.cursor()
    try:
        cursor.execute(insert_req)
        conn.commit()
    except (Exception, psycopg2.DatabaseError) as error:
        print("Error: %s" % error)
        conn.rollback()
        cursor.close()
        return 1
    cursor.close()
```

```python
### Delete data

def delete_data(conn, table_name):
    
    query = f""" TRUNCATE {table_name} CASCADE;; DELETE FROM {table_name}; """

    single_insert(conn, query)
```

```python
delete_data(conn, table_name = 'rol')
delete_data(conn, table_name = 'users')
delete_data(conn, table_name = 'modules')
delete_data(conn, table_name = 'operations')
delete_data(conn, table_name = 'platform')
delete_data(conn, table_name = 'type_shop')
delete_data(conn, table_name = 'shop')
delete_data(conn, table_name = 'launcher')
delete_data(conn, table_name = 'gender_game')
delete_data(conn, table_name = 'franchise')
delete_data(conn, table_name = 'game')
delete_data(conn, table_name = 'rescue')
delete_data(conn, table_name = 'qualification')
```
